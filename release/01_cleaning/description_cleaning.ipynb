```python
import pandas as pd
import re
import os
import csv
```

```python
mnt_path = os.getenv('MNT_PATH') or ''
csv_path = os.path.join(mnt_path, 'complete_data', 'real_and_fake_w_summary.csv')

df = pd.read_csv(
    csv_path,
    dtype=str,
    engine="python",
    sep=",",
    quotechar='"',
    quoting=csv.QUOTE_MINIMAL,
    doublequote=True,
    escapechar=None,
    on_bad_lines="error"
)
```

```python
url_pat = re.compile(r'(https?://\S+|www\.\S+)', flags=re.IGNORECASE)
df['description'] = (
    df['description'].fillna('')
      .str.replace(url_pat, '', regex=True)
      .str.replace(r'\s+', ' ', regex=True)            # 連続空白を1つに
      .str.replace(r'\s+([、。,.!?])', r'\1', regex=True) # 句読点の直前の空白を除去
      .str.strip()
)
```

```python
print(df.columns)
print(len(df))
print(df['is_long'].unique())
```

```python
df.to_csv(
    csv_path,
    index=False,
    sep=",",
    quotechar='"',
    quoting=csv.QUOTE_MINIMAL,
    doublequote=True,
    escapechar=None,
    lineterminator="\n",
    encoding="utf-8",
    na_rep="",
    float_format=None
)
```
